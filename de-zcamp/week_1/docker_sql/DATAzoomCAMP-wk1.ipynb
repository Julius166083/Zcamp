```python
engine=create_engine('postgresql://root:root@localhost:5431/ny_taxi')
```

```python
# this is the statement pandas will execute when creating this table in postgres
print(pd.io.sql.get_schema(df, name= 'yellow_taxi_data_table', con=engine))
```

```python
# to chuck the whole "yellow_tripdata_2021-01" well shall use iterators break it down into batches 
# small chucks of 100,000 then for each chuck is put into the data base until whole scv file is complete

df_iter = pd.read_csv('yellow_tripdata_2021-01.csv', iterator=True, chunksize=100000)
```

```python
# "next" in python returns the next element in the iterator
df=next(df_iter)
```

```python
len(df)
```

```python
# converting date stamps into pandas format and then passing result direct inot the table
df.tpep_pickup_datetime=pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime=pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df.head(n=10)
```

```python
# creating a table with column headings then inserting data thereafter
# if-exists means if table name already exists replace with this one created
# lets do this on column names only and then go to CMD , do"\dt" to see the kind of table we have

df.head(n=0).to_sql(name='yellow_taxi_data_table', con=engine, if_exists='replace')
```

```python
# now insert whole data into the table
# instead of replacing the created table, w'll append now
%time df.to_sql(name='yellow_taxi_data_table', con=engine, if_exists='append')
```

```python
from time import time
```

```python
# iterative code to insert whole of the next dataset,
#when its done, it throws an error "stopiteration"
while True:
    t_start=time()
    
    df=next(df_iter)
    
    df.tpep_pickup_datetime=pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime=pd.to_datetime(df.tpep_dropoff_datetime)
    
    df.to_sql(name='yellow_taxi_data_table', con=engine, if_exists='append')
    
    t_end=time()
    print('inserted another chunck..., took %.3f seconds'%(t_end - t_start))
```
